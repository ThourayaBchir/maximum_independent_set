### Dynamic programming - Maximum independent set for a rooted tree
       
In rooted tree, the maximum independent set is a set of nodes not directly linked with the maximum sum of their weights. Finding this set is a combinatorial optimization problem that can be solved with a recursive algorithm.

<br>

<div>

</div>

```python
class node(object):
    def __init__(self, weight):
        self.weight = weight  # is an integer
        self.children = None  # is a list of node object
        
        
    
```

#### Creating a tree example with the node object:

```python
root = node(3)
```

```python
root.children = [node(4), node(1), node(5)]
```

```python
root.children[0].children = [node(1), node(2)]
```

```python
root.children[1].children = [node(2)]
```

```python
root.children[2].children = [node(1), node(1)]
```

#### Definition of max_weight_independent_set as a recursive function 

```python
def max_weight_independent_set(node):
    
    # if the node is a leaf
    if node.children == None:
        d_node = node.weight
        
    else:
        max1 = node.weight
        for child in node.children:
            grandchildren = child.children
            if grandchildren != None:
                for grandchild in grandchildren:
                    max1 += max_weight_independent_set(grandchild)
        
        max0 = 0
        for child in node.children:
            max0 += max_weight_independent_set(child)
            
        d_node = max(max0, max1)
    
    return d_node
    
    
```

#### Finding the MIS

```python
max_weight_independent_set(root)
```
